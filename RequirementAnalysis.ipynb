```
jamaat = df.groupby('Date').sum()['Cases'].reset_index()
lockdown = dff.groupby('Date').sum()['Cases'].reset_index()
```

```
Jamaat = df.groupby('Date').sum()['Cases'].reset_index()
Lockdown = dff.groupby('Date').sum()['Cases'].reset_index()
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=Jamaat['Date'], y=Jamaat['Cases'], mode='lines+markers', name='Jamaat',line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=Lockdown['Date'], y=Lockdown['Cases'], mode='lines+markers', name='Lockdown', line=dict(color='Red', width=2)))
```
